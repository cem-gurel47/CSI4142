```python
import pandas as pd
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
df = pd.read_csv('/Users/gokhanduzel/Documents/CSI4142/CSI4142/dataset.csv')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.drop_duplicates()
```

We use describe function to summarize our data set.
since our data set is too big, users can look at our whole summarized dataset by opening the printed result within a text editor.

```python
summary = df.describe()
print(summary)
```

```python
avarage_volume = summary.loc['mean', 'volume']
std_volume = summary.loc['std', 'volume']
```

```python
def best_avarages(volume_h):
    if volume_h > avarage_volume:
        return 1
    else:
        return 0
```

```python
df['best_avarages'] = df['volume'].apply(best_avarages)
```

```python
print(df.head())
```

```python
X = df.drop(['company', 'market', 'AVERAGE_SMA_50_VOLUME'], axis=1)
y = df['AVERAGE_SMA_50_VOLUME']
```

```python
selector = SelectKBest(f_classif, k=2)
X_new = selector.fit_transform(X, y)
```

```python
print(X.columns[selector.get_support()])
```
